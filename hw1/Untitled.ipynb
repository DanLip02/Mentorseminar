```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
n = 2 * (10 ** 8)
```

```python
cpp_mpi = np.array([
[1, 0.208394],
[2, 0.139342],
[4, 0.091856],
[8, 0.071856],
[16, 0.057057]
])
cpp_mpi = pd.DataFrame(data=cpp_mpi, columns=["p", "t"])
cpp_mpi["Sp"] = cpp_mpi["t"].iloc[0] / cpp_mpi["t"]
cpp_mpi["Ep"] = cpp_mpi["Sp"] / cpp_mpi["p"]
cpp_mpi["FLOPS"] = (2 * n) / cpp_mpi["t"]
cpp_mpi
```

```python
cpp_omp = np.array([
[1, 0.1900029],
[2, 0.133566],
[4, 0.078199],
[8, 0.051289],
[16, 0.042947]
])
cpp_omp = pd.DataFrame(data=cpp_omp, columns=["p", "t"])
cpp_omp["Sp"] = cpp_omp["t"].iloc[0] / cpp_omp["t"]
cpp_omp["Ep"] = cpp_omp["Sp"] / cpp_omp["p"]
cpp_omp["FLOPS"] = (2 * n) / cpp_omp["t"]
cpp_omp
```

```python
py_omp = np.array([
[1, 0.26243],
[2, 0.143971],
[4, 0.10218],
[8, 0.087703],
[16, 0.071868]
])
py_omp = pd.DataFrame(data=py_omp, columns=["p", "t"])
py_omp["Sp"] = py_omp["t"].iloc[0] / py_omp["t"]
py_omp["Ep"] = py_omp["Sp"] / py_omp["p"]
py_omp["FLOPS"] = (2 * n) / py_omp["t"]
py_omp
```

```python
py_mpi = np.array([
[1, 38.574857],
[2, 19.068666],
[4, 9.507275],
[8, 4.653888],
[16, 3.384730]
])
py_mpi = pd.DataFrame(data=py_mpi, columns=["p", "t"])
py_mpi["Sp"] = py_mpi["t"].iloc[0] / py_mpi["t"]
py_mpi["Ep"] = py_mpi["Sp"] / py_mpi["p"]
py_mpi["FLOPS"] = (2 * n) / py_mpi["t"]
py_mpi
```

```python
col = "t"

plt.figure(figsize=(10, 6))
plt.plot(py_omp["p"], py_omp[col], marker='o', linestyle='--', label='py_omp')
plt.plot(cpp_omp["p"], cpp_omp[col], marker='o', linestyle='--', label='cpp_omp')
plt.plot(py_mpi["p"], py_mpi[col], marker='o', linestyle='--', label='py_mpi')
plt.plot(cpp_mpi["p"], cpp_mpi[col], marker='o', linestyle='--', label='cpp_mpi')
plt.xlabel("p")
plt.ylabel("T")
plt.title("Tp")
plt.grid(True)
plt.legend()
plt.show()
```
